```python
%matplotlib inline

import feedparser

from pprint import pprint
from textwrap import TextWrapper

feed_url = "http://mashable.com/feed/"
feeds = feedparser.parse(feed_url)
print('\nNo. of feeds: ' + str(len(feeds)))
print('Raw Feed as JSON: \n')
pprint(feeds)
```

```python
tags = []
for i, entry in enumerate(feeds['entries']):
    print("\nEntry No. :- " + str(i + 1))
    print("\n\tTitle :- " + entry['title'])
    print("\tPost Url :- " + entry['link'])
    num_tags = len(entry['tags'])
    print("\tNo. of tags :- " + str(num_tags))
    
    wrapper = TextWrapper(initial_indent='\n\tSummary: ', width=80,
                               subsequent_indent=' ' * len('Summary:'))
    
    print(wrapper.fill(entry['summary']))
    
    for i in range(num_tags):
        term = entry['tags'][i]['term']
        if " " in term:
            tags.append("".join(term.split()))
        else:
            tags.append(term)
```

```python
tags
```

### Now lets get the frequency of each tag with their duplicate counts


```python
from collections import Counter

counter = Counter(tags)
for key, value in counter.items():
    print(key + " - " + str(value))
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

#counter = sorted(counter, key=counter.get, reverse=True) # Sort the dictionary by value
top_10 = dict(sorted(counter.items(), key=lambda x: x[1])[-10:])
tag_names = top_10.keys()
y_pos = np.arange(len(tag_names))
freq = top_10.values()
 
plt.barh(y_pos, freq, align='center', alpha=0.8, color='green')
plt.yticks(y_pos, tag_names)
plt.xlabel('Frequency Counts')
plt.title('Top 10 Tag Frequency BarChart')
 
plt.show()
```
